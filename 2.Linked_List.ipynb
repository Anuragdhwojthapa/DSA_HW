# LINKED LIST

```python
#1. Write a program to show the implementation of Linked List in python.
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def __repr__(self):
        nodes = []
        temp = self.head
        while temp:
            nodes.append(str(temp.data))
            temp = temp.next
        return " -> ".join(nodes) + " -> None"

# Example
n1 = Node(11)
n2 = Node(22)
n3 = Node(33)

n1.next = n2
n2.next = n3

lst = LinkedList()
lst.head = n1
print(lst)   
```

```python
#2. Write the implementation of Singly Linked Lists to show creation, traversal, insertion, retrieval and deletion of elements. 
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def __repr__(self):
        if self.head is None:
            return "List is empty"
        nodes = []
        temp = self.head
        while temp:
            nodes.append(str(temp.data))
            temp = temp.next
        return " -> ".join(nodes) + " -> None"

    def insert_at_beginning(self, value):
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node

    def insert_at_end(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                return True
            temp = temp.next
        return False

    def delete(self, key):
        temp = self.head
        prev = None
        while temp:
            if temp.data == key:
                if prev:
                    prev.next = temp.next
                else:
                    self.head = temp.next
                return
            prev = temp
            temp = temp.next

# Example
sll = SinglyLinkedList()
sll.insert_at_beginning(6)
sll.insert_at_end(7)
sll.insert_at_end(65)
print(sll)
print("Search 20:", sll.search(20))
sll.delete(20)
print("After deletion:", sll)
```

```python
#3. Write a program to implement the Doubly Linked List. 
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def __repr__(self):
        if self.head is None:
            return "List is empty"
        nodes = []
        temp = self.head
        while temp:
            nodes.append(str(temp.data))
            temp = temp.next
        return " <-> ".join(nodes) + " <-> None"

    def insert_at_end(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp

# Example
dll = DoublyLinkedList()
dll.insert_at_end(100)
dll.insert_at_end(200)
dll.insert_at_end(300)
print(dll)
```

```python
#4. Write a program to implement Circular Lists. 
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def __repr__(self):
        if self.head is None:
            return "List is empty"
        nodes = []
        temp = self.head
        while True:
            nodes.append(str(temp.data))
            temp = temp.next
            if temp == self.head:
                break
        return " -> ".join(nodes) + " -> (back to head)"

    def insert_at_end(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            return
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = new_node
        new_node.next = self.head

# Example
cll = CircularLinkedList()
cll.insert_at_end(12)
cll.insert_at_end(67)
cll.insert_at_end(38)
print(cll)
```

```python
#5. Show the implementation of stack and queue using the Linked List created above. 
# Stack using Singly Linked List
class Stack:
    def __init__(self):
        self.sll = SinglyLinkedList()

    def push(self, value):
        self.sll.insert_at_beginning(value)

    def pop(self):
        if self.sll.head is None:
            return "Stack is empty"
        value = self.sll.head.data
        self.sll.head = self.sll.head.next
        return value

    def peek(self):
        return None if self.sll.head is None else self.sll.head.data

# Queue using Singly Linked List
class Queue:
    def __init__(self):
        self.sll = SinglyLinkedList()

    def enqueue(self, value):
        self.sll.insert_at_end(value)

    def dequeue(self):
        if self.sll.head is None:
            return "Queue is empty"
        value = self.sll.head.data
        self.sll.head = self.sll.head.next
        return value

# Example
stack = Stack()
stack.push(10)
stack.push(20)
print("Stack top:", stack.peek())
print("Popped:", stack.pop())

queue = Queue()
queue.enqueue(10)
queue.enqueue(20)
print("Dequeued:", queue.dequeue())
```
