```python
# DFS Traversal

class Graph:
    def __init__(self):
        self.adj = {} 

    def add_vertex(self, vertex):
        if vertex not in self.adj:
            self.adj[vertex] = []

    def Add_Edge_udg(self, src, dest, weight=1):
        self.add_vertex(src)
        self.add_vertex(dest)
        if dest not in self.adj[src]:
            self.adj[src].append(dest)
        if src not in self.adj[dest]:
            self.adj[dest].append(src)

    def DFS(self, src):
        if src not in self.adj:
            print(f"Vertex '{src}' not found in graph.")
            return

        visited = set()
        stack = [src]

        print("DFS Traversal:")
        while stack:
            vertex = stack.pop()
            if vertex not in visited:
                print(vertex, end=" --> ")
                visited.add(vertex)

                # Push unvisited neighbors onto the stack
                for neighbor in self.adj[vertex]:
                    if neighbor not in visited:
                        stack.append(neighbor)

        print("END")


# Graph Structure 

#       I --- H   G      E
#      /       \ /      /
#     A ------- B ---- C ---- D
#                      \
#                       F

graph = Graph()

graph.Add_Edge_udg('A', 'B')  
graph.Add_Edge_udg('A', 'I')   
graph.Add_Edge_udg('B', 'G')   
graph.Add_Edge_udg('B', 'H')   
graph.Add_Edge_udg('B', 'C')   
graph.Add_Edge_udg('C', 'E')   
graph.Add_Edge_udg('C', 'D')   
graph.Add_Edge_udg('C', 'F')   
graph.Add_Edge_udg('I', 'H')   

graph.DFS('A')
```

```python
# BFS Traversal

from collections import deque

class Graph:
    def __init__(self):
        self.adj = {}  

    def add_vertex(self, vertex):
        if vertex not in self.adj:
            self.adj[vertex] = []

    def Add_Edge_udg(self, src, dest):
        self.add_vertex(src)
        self.add_vertex(dest)
        if dest not in self.adj[src]:
            self.adj[src].append(dest)
        if src not in self.adj[dest]:
            self.adj[dest].append(src)

    def BFS(self, src):
        if src not in self.adj:
            print(f"Vertex '{src}' not found in graph.")
            return

        visited = set()
        queue = deque([src])       
        visited.add(src)           

        print("BFS Traversal:")
        while queue:
            vertex = queue.popleft()           
            print(vertex, end=" --> ")

            for neighbor in self.adj[vertex]:
                if neighbor not in visited:
                    visited.add(neighbor)      
                    queue.append(neighbor)     
        print("END")


# Graph Structure 

#       I --- H   G      E
#      /       \ /      /
#     A ------- B ---- C ---- D
#                      \
#                       F

graph = Graph()

graph.Add_Edge_udg('A', 'B')   
graph.Add_Edge_udg('A', 'I')   
graph.Add_Edge_udg('B', 'G')   
graph.Add_Edge_udg('B', 'H')   
graph.Add_Edge_udg('B', 'C')   
graph.Add_Edge_udg('C', 'E')   
graph.Add_Edge_udg('C', 'D')   
graph.Add_Edge_udg('C', 'F')   
graph.Add_Edge_udg('I', 'H')  

graph.BFS('A')
```

```python
# 2. Write a program to implement Kruskal's Algorithm for minimum spanning tree.
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])  # Path compression
        return self.parent[x]

    def union(self, x, y):
        px, py = self.find(x), self.find(y)
        if px == py:
            return False  # Already in the same component (cycle)
        # Union by rank
        if self.rank[px] < self.rank[py]:
            px, py = py, px
        self.parent[py] = px
        if self.rank[px] == self.rank[py]:
            self.rank[px] += 1
        return True


def kruskal(n, edges):
    """
    n     : number of vertices (0 to n-1)
    edges : list of (weight, u, v)
    returns: (MST cost, list of edges in MST)
    """
    edges.sort()  # Sort edges by weight
    uf = UnionFind(n)
    mst_cost = 0
    mst_edges = []

    for weight, u, v in edges:
        if uf.union(u, v):          # Add edge only if it doesn't form a cycle
            mst_cost += weight
            mst_edges.append((u, v, weight))
            if len(mst_edges) == n - 1:  # MST is complete
                break

    return mst_cost, mst_edges


# Example 
if __name__ == "__main__":
    n = 6
    edges = [
        (4, 0, 1),
        (2, 0, 2),
        (6, 1, 2),
        (5, 1, 3),
        (3, 2, 3),
        (8, 2, 4),
        (7, 3, 4),
        (1, 3, 5),
        (9, 4, 5),
    ]

    cost, mst = kruskal(n, edges)

    print(f"MST Total Cost: {cost}")
    print("Edges in MST:")
    for u, v, w in mst:
        print(f"  {u} -- {v}  (weight {w})")
```

```python
# 3. Write a program to implement Prim's Algorithm to illustrate minimum spanning tree.
import heapq
from collections import defaultdict

def prim(n, edges, start=0):
    """
    n      : number of vertices (0 to n-1)
    edges  : list of (u, v, weight)
    start  : starting vertex
    returns: (MST cost, list of edges in MST)
    """
    # Build adjacency list
    graph = defaultdict(list)
    for u, v, w in edges:
        graph[u].append((w, v))
        graph[v].append((w, u))

    visited = set()
    min_heap = [(0, start, -1)]  # (weight, current_node, parent_node)
    mst_cost = 0
    mst_edges = []

    while min_heap and len(visited) < n:
        weight, u, parent = heapq.heappop(min_heap)

        if u in visited:
            continue

        visited.add(u)
        mst_cost += weight

        if parent != -1:
            mst_edges.append((parent, u, weight))

        for edge_weight, v in graph[u]:
            if v not in visited:
                heapq.heappush(min_heap, (edge_weight, v, u))

    return mst_cost, mst_edges


# Example 
if __name__ == "__main__":
    n = 6
    edges = [
        (0, 1, 4),
        (0, 2, 2),
        (1, 2, 6),
        (1, 3, 5),
        (2, 3, 3),
        (2, 4, 8),
        (3, 4, 7),
        (3, 5, 1),
        (4, 5, 9),
    ]

    cost, mst = prim(n, edges, start=0)

    print(f"MST Total Cost: {cost}")
    print("Edges in MST:")
    for u, v, w in mst:
        print(f"  {u} -- {v}  (weight {w})")
```
