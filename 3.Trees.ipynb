```python
#2. Implement BFS for value retrieval from tree in python.
from collections import deque

class Node:
    """A node in a binary tree."""
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    def __repr__(self):
        return str(self.data)


class BFS_Traversal:
    """Breadth-First Search (BFS) / Level Order Traversal."""

    def __init__(self, root):
        self.root = root

    def traverse(self):
        if self.root is None:
            return

        queue = deque()
        queue.append(self.root)

        while queue:
            node = queue.popleft()
            print(node.data, end=" ")

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)


if __name__ == "__main__":
    # Construct the tree
    root = Node(10)
    root.left = Node(20)
    root.right = Node(30)
    root.left.left = Node(40)
    root.left.right = Node(50)
    root.right.left = Node(60)
    root.right.right = Node(70)

    print("Binary Tree BFS / Level Order Traversal:")
    bfs = BFS_Traversal(root)
    bfs.traverse()
```

```python
#3. Implement DFS for value retrieval from tree in python.

class Node:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None

    def __repr__(self):
        return str(self.data)    

# Preorder Traversal (Root → Left → Right)
def Preorder(root):
    if root != None:
        print(root.data , end = " --> ")
        Preorder(root.left)
        Preorder(root.right)

# Inorder Traversal (Left → Root → Right)
def Inorder(root):
    if root != None:
        Inorder(root.left)
        print(root.data , end = " --> ")
        Inorder(root.right)

# Postorder Traversal (Left → Right → Root)
def Postorder(root):
    if root != None:
        Postorder(root.left)
        Postorder(root.right)
        print(root.data , end = " --> ")

# Tree
root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.left.right = Node(50)
root.right.left = Node(60)
root.right.right = Node(70)

print("Preorder Traversal:",end = " ")
Preorder(root)
print("\nInorder Traversal:",end = " ")
Inorder(root)
print("\nPostorder Traversal:",end = " ")
Postorder(root)
```

```python
#4. Write a program in python to implement Binary Search tree.
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def insert(root, value):
    if root is None:
        return TreeNode(value)
    if value < root.data:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)
    return root

def search(root, value):
    if root is None:
        return False
    if root.data == value:
        return True
    elif value < root.data:
        return search(root.left, value)
    else:
        return search(root.right, value)

def inorder(root):
    if root:
        inorder(root.left)
        print(root.data, end=" ")
        inorder(root.right)

# Example
root = None
nums = [45, 20, 65, 10, 30, 55, 70]
for n in nums:
    root = insert(root, n)

print("Inorder Traversal of BST:")
inorder(root)
print("\nSearch 30:", search(root, 30))
print("Search 99:", search(root, 99))
```

```python
#5. Implement AVL Tree in python.
class AVLNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        self.height = 1

def height(node):
    return node.height if node else 0

def balance(node):
    return height(node.left) - height(node.right) if node else 0

def rotate_right(y):
    x = y.left
    T2 = x.right
    x.right = y
    y.left = T2
    y.height = 1 + max(height(y.left), height(y.right))
    x.height = 1 + max(height(x.left), height(x.right))
    return x

def rotate_left(x):
    y = x.right
    T2 = y.left
    y.left = x
    x.right = T2
    x.height = 1 + max(height(x.left), height(x.right))
    y.height = 1 + max(height(y.left), height(y.right))
    return y

def insert(node, key):
    if not node:
        return AVLNode(key)
    if key < node.data:
        node.left = insert(node.left, key)
    else:
        node.right = insert(node.right, key)

    node.height = 1 + max(height(node.left), height(node.right))
    bf = balance(node)

    if bf > 1 and key < node.left.data:
        return rotate_right(node)
    if bf < -1 and key > node.right.data:
        return rotate_left(node)
    if bf > 1 and key > node.left.data:
        node.left = rotate_left(node.left)
        return rotate_right(node)
    if bf < -1 and key < node.right.data:
        node.right = rotate_right(node.right)
        return rotate_left(node)

    return node

def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=" ")
        inorder(node.right)

# Example
root = None
for val in [25, 15, 35, 10, 20, 30, 40]:
    root = insert(root, val)

print("AVL Tree Inorder Traversal:")
inorder(root)
```

```python
#6. Write a program to implement Heap data structure in python.
import heapq

class MinHeap:
    def __init__(self):
        self.heap = []

    def insert(self, val):
        heapq.heappush(self.heap, val)

    def remove(self):
        if not self.heap:
            return "Heap is empty"
        return heapq.heappop(self.heap)

# Example
h = MinHeap()
for n in [12, 5, 18, 3, 25]:
    h.insert(n)

print("Heap:", h.heap)
print("Removed:", h.remove())
print("Heap after removal:", h.heap)
```

```python
#7. Write a program to implement priority queue using heap data structure in python.
import heapq

class PriorityQueue:
    def __init__(self):
        self.queue = []

    def insert(self, item, priority):
        heapq.heappush(self.queue, (-priority, item))  # negative for max-heap

    def delete(self):
        if not self.queue:
            return "Queue is empty"
        priority, item = heapq.heappop(self.queue)
        return f"Deleted {item} with priority {-priority}"

# Example
pq = PriorityQueue()
pq.insert("TaskA", 3)
pq.insert("TaskB", 5)
pq.insert("TaskC", 1)

print(pq.delete())
print(pq.delete())
```
