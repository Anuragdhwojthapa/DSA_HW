```python
#1. Write a program to show the implementation of sequential Search.
def sequential_search(arr, key):
    # Travering the list one by one
    for i in range(len(arr)):
        if arr[i] == key:
            return f"Element {key} found at index {i}"
    return f"Element {key} not found"

# Example
arr = [12, 45, 23, 67, 89, 34]
print(sequential_search(arr, 67))
print(sequential_search(arr, 100))
```

```python
#2. Write the implementation of Binary Search in Python. 
# Binary Search (array must be sorted)
def binary_search(arr, key):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return f"Element {key} found at index {mid}"
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return f"Element {key} not found"

# Example
arr = [10, 20, 30, 40, 50, 60, 70]
print(binary_search(arr, 50))
print(binary_search(arr, 25))
```

```python
#3. Show the basic implementation of Hashing in python 

class HashTable:
    def __init__(self):
        self.table = {}

    def insert(self, key, value):
        self.table[key] = value

    def search(self, key):
        return self.table.get(key, "Key not found")

# Example
ht = HashTable()
ht.insert(1, "Apple")
ht.insert(2, "Banana")
ht.insert(3, "Cherry")

print(ht.search(2))
print(ht.search(5))
```

```python
#4. Write a program to show linear, quadratic probing and double hashing in python.
class HashTableProbing:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_func(self, key):
        return key % self.size

    def insert_linear(self, key):
        index = self.hash_func(key)
        while self.table[index] is not None:
            index = (index + 1) % self.size
        self.table[index] = key

    def insert_quadratic(self, key):
        index = self.hash_func(key)
        i = 1
        while self.table[index] is not None:
            index = (index + i*i) % self.size
            i += 1
        self.table[index] = key

    def insert_double_hash(self, key):
        index = self.hash_func(key)
        step = 7 - (key % 7)   # second hash function
        while self.table[index] is not None:
            index = (index + step) % self.size
        self.table[index] = key

# Example
ht = HashTableProbing(10)
ht.insert_linear(23)
ht.insert_linear(33)
ht.insert_quadratic(44)
ht.insert_double_hash(55)
print("Hash Table:", ht.table)
```

```python
#5. Write a program to implement Separate Chaining in python.
class HashTableChaining:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_func(self, key):
        return key % self.size

    def insert(self, key, value):
        index = self.hash_func(key)
        self.table[index].append((key, value))

    def search(self, key):
        index = self.hash_func(key)
        for k, v in self.table[index]:
            if k == key:
                return v
        return "Key not found"

# Example
ht = HashTableChaining(5)
ht.insert(10, "Apple")
ht.insert(15, "Banana")
ht.insert(20, "Cherry")
print(ht.search(15))
print(ht.search(25))
```
