## Stacks

```python
#1. Write a program to implement stack from scratch in python using array.
class ArrayStack:

    def __init__(self, length):
        self.length = length
        self.s = [None]*length
        self.count = 0
    def push(self, data):
        if self.count == self.length:
            print("Stack is full")
            return
        self.s[self.count] = data
        self.count += 1
    def pop(self):
        if self.count == 0:
            print("Stack is empty")
            return
        self.count -= 1
        return self.s[self.count]
    def peek(self):
        if self.count == 0:
            print("Stack is empty")
            return
        return self.s[self.count-1]
    def is_empty(self):
        return self.count == 0
    def size(self):
        return self.count

# Example    
stack = ArrayStack(15)
stack.push(100)  
stack.push(222)
print(stack.pop())
print(stack.peek())
print(stack.is_empty())
print(stack.size())
```

```python
#2. Write a program in python to implement stack using python list.
class ListStack:

    def __init__(self):
        self.s = []
    def push(self, data):
        self.s.append(data)
    def pop(self):
        if len(self.s) == 0:
            print("Stack is empty")
            return
        return self.s.pop()
    def peek(self):
        if len(self.s) == 0:
            print("Stack is empty")
            return
        return self.s[-1]
    def is_empty(self):
        return len(self.s) == 0
    def size(self):
        return len(self.s)

# Example    
stack = ListStack()
stack.push(11)
stack.push(22)
print(stack.pop())
print(stack.peek())
print(stack.is_empty())
print(stack.size())
```

```python
#3. Write a program for Bracket Matching using using above created Stack.
def is_balanced(s):
    stack = []
    opening = set('([{')
    closing = set(')]}')
    mapping = {')': '(', ']': '[', '}': '{'}
    for char in s:
        if char in opening:
            stack.append(char)
        elif char in closing:
            if not stack or stack[-1] != mapping[char]:
                return False
            stack.pop()
    return not stack

# Example
print(is_balanced("()[{}]"))
print(is_balanced("([)]"))
```

```python
#4. Write a program in python to implement Infix to Postfix conversion using Stack implemented above.
def infix_to_postfix(expression):
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}
    output = []
    stack = []
    for char in expression:
        if char.isalnum():
            output.append(char)
        elif char in precedence:
            while (stack and stack[-1] != '(' and
                   precedence[stack[-1]] >= precedence[char]):
                output.append(stack.pop())
            stack.append(char)
        elif char == '(':
            stack.append(char)
        elif char == ')':
            while stack and stack[-1] != '(':
                output.append(stack.pop())
            stack.pop()  
    while stack:
        output.append(stack.pop())
    return ''.join(output)

# Example
print(infix_to_postfix("A+B*C"))
print(infix_to_postfix("(A+B)*C"))  
```

```python
#5. Write a program in python to evluate postfix expression obtained from above implementation.
def evaluate_postfix(expression):
    stack = []
    for char in expression:
        if char.isdigit():
            stack.append(int(char))
        else:
            b = stack.pop()
            a = stack.pop()
            if char == '+':
                stack.append(a + b)
            elif char == '-':
                stack.append(a - b)
            elif char == '*':
                stack.append(a * b)
            elif char == '/':
                stack.append(a / b)
    return stack[0]

# Example
print(evaluate_postfix("11+22*"))
print(evaluate_postfix("1122+*"))
```

## Queues

```python
#6. Implement a queue using array in python.
class ArrayQueue:

    def __init__(self, length):
        self.length = length
        self.q = [None]*length
        self.front = 0
        self.rear = 0
        self.count = 0
    def enqueue(self, data):
        if self.count == self.length:
            print("Queue is full")
            return
        self.q[self.rear] = data
        self.rear = (self.rear + 1) % self.length
        self.count += 1
    def dequeue(self):
        if self.count == 0:
            print("Queue is empty")
            return
        data = self.q[self.front]
        self.front = (self.front + 1) % self.length
        self.count -= 1
        return data
    def peek(self):
        if self.count == 0:
            print("Queue is empty")
            return
        return self.q[self.front]
    def is_empty(self):
        return self.count == 0
    def size(self):
        return self.count

#Example
queue = ArrayQueue(10)
queue.enqueue(11)
queue.enqueue(22)
print(queue.dequeue())
```

```python
#7. Implement a priority queue using array in python.
class Node:
    def __init__(self, info, priority):
        self.info = info
        self.priority = priority 

class PriorityQueue:
    def __init__(self):
        self.queue = []
    def size(self):
        return len(self.queue)
    def show(self):
        print("Priority Queue:")
        for x in self.queue:
            print(f"{x.info} - {x.priority}")
    def insert(self, node):
        """Insert node based on priority (higher number = higher priority)"""
        # If queue is empty, just append
        if not self.queue:
            self.queue.append(node)
            return
        # Find the position to insert
        for i, current in enumerate(self.queue):
            if node.priority > current.priority:
                self.queue.insert(i, node)
                return
        # If not inserted yet, append at the end
        self.queue.append(node)
    def delete(self):
        if not self.queue:
            print("Queue is empty!")
            return None
        x = self.queue.pop(0)
        print(f"Deleted data with priority {x.priority}: {x.info}")
        return x

# Example 
if __name__ == "__main__":
    p = PriorityQueue()
    p.insert(Node("Ram", 16))
    p.insert(Node("Sam", 12))
    p.insert(Node("John", 1))
    p.insert(Node("Don", 25))
    p.insert(Node("Monk", 26))
    
    p.show()
    p.delete()
    print("\nAfter deletion:")
    p.show()
```
